```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('placement.csv')
```

```python
df.head()
```

```python
plt.scatter(df['cgpa'],df['package'])
plt.xlabel('CGPA')
plt.ylabel('Package(in lpa)')
```

```python
X = df.iloc[:,0:1]
# X = df.iloc[:,0]   -> returns a Series.
# X = df.iloc[:,0:1]  -> returns a DataFrame.
# Most machine learning models expect X to be a 2D structure (DataFrame), so 0:1 is commonly used

y = df.iloc[:,-1]
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
X_test
```

```python
y_test
```

```python
lr.predict(X_test.iloc[0].values.reshape(1,1))   
# Takes the first test sample, converts it into the 2D format required by scikit-learn, 
# and predicts its output using the trained linear regression model.
```

```python
plt.scatter(df['cgpa'],df['package'])
plt.plot(X_train,lr.predict(X_train),color='red')
plt.xlabel('CGPA') 
plt.ylabel('Package(in lpa)')
```

```python
m = lr.coef_
```

```python
b = lr.intercept_
```
