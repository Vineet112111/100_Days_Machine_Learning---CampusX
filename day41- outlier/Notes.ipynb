# OUTLIERS IN MACHINE LEARNING - COMPLETE NOTES

## 1. Introduction to Outliers

Outliers are one of the most important concepts in Feature Engineering and Data Preprocessing.

Before training any Machine Learning model, data must be analyzed carefully because abnormal values can significantly affect model performance.

An outlier is a data point that is very different from the majority of observations in a dataset.

### Example

Salary:
25000
30000
28000
35000
50000000

Here, 50000000 is an outlier because it is far away from the rest of the values.

---

## 2. Formal Definition of Outlier

An outlier is an observation that lies at an abnormal distance from other observations in a dataset.

Outlier = Value significantly different from the majority of data

Outliers may occur because of:

1. Human errors
2. Measurement errors
3. Data entry mistakes
4. Natural variation
5. Rare events

---

## 3. Why Do Outliers Matter?

Outliers influence:

- Mean
- Variance
- Standard Deviation
- Correlation
- Machine Learning Models

They can completely change patterns learned by a model.

---

## 4. Example of Outlier Impact on Mean

Dataset:

10, 12, 15, 18, 20

Mean:

15

Add Outlier:

10, 12, 15, 18, 20, 100

New Mean:

29.17

The entire average shifts because of one value.

This demonstrates how powerful outliers can be.

---

## 5. Types of Outliers

### A. Error Outliers

These are generated because of mistakes.

Example:

Age = 850

Human age cannot be 850.

This is an invalid observation.

Action:
Remove or correct.

---

### B. Genuine Outliers

These values are unusual but correct.

Example:

Most salaries:
₹30,000 – ₹2,00,000

Mukesh Ambani's salary:
Several Crores

Looks unusual but is valid.

Action:
Keep the observation.

---

### C. Contextual Outliers

These values appear abnormal because some information is missing.

Example:

Student studies:
1 hour/day

Score:
99%

Looks impossible.

But maybe:

- High IQ
- Previous preparation
- Coaching

The issue is not the outlier.

The issue is missing features.

---

## 6. When Are Outliers Dangerous?

This is one of the most important concepts in Machine Learning.

### Case 1: Data Entry Errors

Example:

Age = 700

Impossible.

These values create noise.

Solution:
Remove them.

---

### Case 2: Weight-Based Algorithms

Outliers are extremely dangerous for:

- Linear Regression
- Logistic Regression
- Neural Networks
- Deep Learning

These algorithms learn by minimizing error.

Outliers create huge errors.

Huge errors force the model to change weights aggressively.

Result:

- Poor Predictions
- Overfitting
- Low Accuracy

---

### Case 3: Small Dataset

Suppose dataset contains:

20 rows

Outliers:

3 rows

Now:

15% of data is abnormal

The model becomes highly biased.

---

### Case 4: Distance-Based Algorithms

Algorithms:

- KNN
- K-Means Clustering

depend on distance calculations.

Outliers increase distances dramatically.

Clusters become incorrect.

Nearest neighbors become misleading.

---

## 7. When Should We NOT Remove Outliers?

Many beginners remove all outliers.

This is wrong.

### Fraud Detection

Normal Transactions:

₹100
₹200
₹500

Fraud Transaction:

₹5,00,000

Fraud itself is an outlier.

Removing it destroys useful information.

---

### Medical Diagnosis

Rare diseases often appear as outliers.

Removing them may remove important medical cases.

---

### Cyber Security

Hackers behave differently from normal users.

Their activity appears as outliers.

Removing them removes attack signals.

---

### Stock Market Crashes

Market crashes are rare.

They appear as outliers.

Yet they are extremely important.

---

### Golden Rule

Before removing any outlier ask:

Is it an error?
OR
Is it valuable information?

---

## 8. How Outliers Affect Machine Learning Models

Suppose we have:

Study Hours | Marks
1           | 20
2           | 40
3           | 60
4           | 80

Perfect Linear Relationship:

Marks = 20 × Hours

Add Outlier:

Study Hours | Marks
1           | 100

Now regression line shifts.

Model learns wrong pattern.

This phenomenon is called:

Model Distortion

---

## 9. Outlier Treatment Techniques

### Method 1: Trimming

Definition:

Removing rows containing outliers.

Example:

Before:

20
30
40
500

After:

20
30
40

500 removed.

Advantages:

- Easy
- Fast
- Effective

Disadvantages:

- Data loss
- Important information may disappear

---

### Method 2: Capping

Also known as:

Winsorization

Definition:

Replace extreme values with predefined limits.

Example:

Upper Limit = 100

Data:

20
30
50
200
500

After Capping:

20
30
50
100
100

Types of Capping:

1. Upper Capping
2. Lower Capping
3. Both Side Capping

Advantages:

- No data loss
- Preserves dataset size

Disadvantages:

- Some information lost

---

### Method 3: Missing Value Treatment

Convert outliers into NaN.

Example:

Before:

20
30
40
500

After:

20
30
40
NaN

Then apply:

- Mean Imputation
- Median Imputation
- KNN Imputation

Advantages:

Allows use of existing missing-value techniques.

Disadvantages:

May lose information.

---

### Method 4: Discretization (Binning)

Convert numerical values into categories.

Example:

Age:

10
15
20
25
80

After Binning:

10 → Child
15 → Teen
20 → Adult
25 → Adult
80 → Senior

Now the impact of 80 reduces.

Advantages:

- Reduces influence of outliers

Disadvantages:

- Information loss

---

## Can Binning Fix Outliers?

Answer:

Partially

Binning does not remove outliers.

It only reduces their influence.

---

## 10. Detecting Outliers

Three methods are introduced.

### Method 1: Z-Score Method

Best For:

Normally Distributed Data

Formula:

Z = (X - Mean) / Standard Deviation

Rule:

|Z| > 3

Then:

Outlier

Example:

Mean = 50
Standard Deviation = 10
Value = 85

Z = (85 - 50) / 10

Z = 3.5

Since:

3.5 > 3

Outlier.

Limitation:

Works well only for normal distributions.

---

### Method 2: IQR Method

Best for:

Skewed Data

Step 1:

Q1 = 25th Percentile
Q3 = 75th Percentile

Step 2:

IQR = Q3 - Q1

Step 3:

Lower Bound = Q1 - 1.5 × IQR

Upper Bound = Q3 + 1.5 × IQR

Example:

Q1 = 20
Q3 = 40

IQR = 20

Lower Bound:

20 - 1.5(20) = -10

Upper Bound:

40 + 1.5(20) = 70

Any value:

< -10
OR
> 70

is an outlier.

Why IQR for Skewed Data?

Mean and Standard Deviation become unreliable in skewed distributions.

Quartiles remain stable.

Therefore:

IQR = Best Choice for Skewed Data

---

### Method 3: Percentile Method

Very flexible.

Example:

Choose:

1st Percentile
99th Percentile

Suppose:

P1 = 10
P99 = 200

Then:

Values < 10
OR
Values > 200

are considered outliers.

Advantages:

- Works with any distribution
- Easy to implement
- Widely used in industry

---

## 11. Winsorization

Winsorization is an advanced form of capping.

Example:

99th Percentile = 150

Values:

200
300
500

Become:

150
150
150

No row is deleted.

Dataset size remains unchanged.

---

## 12. Model Sensitivity to Outliers

### Highly Sensitive Models

- Linear Regression
- Logistic Regression
- Neural Networks
- Deep Learning
- KNN
- K-Means

### Less Sensitive Models

- Decision Trees
- Random Forest
- XGBoost
- CatBoost
- LightGBM

Because they split data rather than fitting global lines.

## Interview Questions

Q. What is an outlier?

A. A data point significantly different from the majority of observations.

---

Q. Why are outliers dangerous?

A. They distort statistics and machine learning models.

---

Q. Why shouldn't all outliers be removed?

A. Some outliers represent valuable information such as fraud, diseases, or rare events.

---

Q. What is trimming?

A. Removing rows containing outliers.

---

Q. What is capping?

A. Replacing extreme values with predefined threshold values.

---

Q. What is Winsorization?

A. A capping technique that replaces extreme values with percentile limits.

---

Q. Which method is best for normal data?

A. Z-Score Method.

---

Q. Which method is best for skewed data?

A. IQR Method.

---

Q. Which method is most flexible?

A. Percentile Method.
