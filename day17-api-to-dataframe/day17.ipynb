# Faster Version

```python
# Loop through all pages of movie data
for i in range(1, 429):

    # Send request to TMDB API
    response = requests.get(f'https://api.themoviedb.org/3/movie/top_rated?api_key=d0dfd261b932115b6f3b202ad86004fd&language=en-US&page={i}')

    # Convert JSON data to DataFrame
    temp_df = pd.DataFrame(response.json()['results'])

    # Keep only required columns
    temp_df = temp_df[['id','title','overview',
                       'release_date','popularity',
                       'vote_average','vote_count']]

    # Add current page data to main DataFrame
    df = pd.concat([df, temp_df], ignore_index=True)
```
