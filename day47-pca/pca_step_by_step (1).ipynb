# Comprehensive Technical Notes: PCA (Problem Formulation & Solution)

## 1. Problem Formulation (Mathematical Objective)

* **Goal:** Reduce data from a $d$-dimensional space to a $k$-dimensional space ($k < d$) while maximizing the variance of the projected data.
* **Input Data Point:** Represented as a vector $x_i \in \mathbb{R}^d$.
* **Projection Axis:** Represented as a unit vector $u \in \mathbb{R}^d$, where the magnitude constraint is:

$$\|u\| = 1 \implies u^T u = 1$$


* **Projection Equation:** The scalar projection of an original data point $x_i$ onto the unit vector $u$ is given by the dot product:

$$x_i' = u^T x_i$$


* **Projected Mean:** If the mean of the original data is $\bar{x}$, the mean of the projected data is:

$$\bar{x}' = u^T \bar{x}$$



### Optimization Objective

Maximize the variance of the projected points ($x_i'$):


$$\text{Maximize } \frac{1}{n} \sum_{i=1}^{n} (u^T x_i - u^T \bar{x})^2$$


Subject to the constraint:


$$u^T u = 1$$

---

## 2. Statistical Prerequisites

### Covariance

* **Formula for two variables $X$ and $Y$:**

$$\text{Cov}(X, Y) = \frac{1}{n}\sum_{i=1}^{n}(x_i - \bar{x})(y_i - \bar{y})$$


* **Property:** $\text{Cov}(X, Y) = \text{Cov}(Y, X)$

### Covariance Matrix ($\Sigma$)

For a 3D space with features $F_1, F_2, F_3$, the covariance matrix is a symmetric $3 \times 3$ matrix:

$$\Sigma = \begin{bmatrix} 
\text{Var}(F_1) & \text{Cov}(F_1, F_2) & \text{Cov}(F_1, F_3) \\ 
\text{Cov}(F_2, F_1) & \text{Var}(F_2) & \text{Cov}(F_2, F_3) \\ 
\text{Cov}(F_3, F_1) & \text{Cov}(F_3, F_2) & \text{Var}(F_3) 
\end{bmatrix}$$

* **Diagonal elements:** Variances of individual features.
* **Off-diagonal elements:** Covariances between feature pairs.

---

## 3. Linear Algebra Prerequisites

### Matrix Transformations

* Multiplying a matrix $A$ by a vector $x$ transforms the vector into a new vector $x'$:

$$A x = x'$$



### Eigenvectors and Eigenvalues

* **Fundamental Equation:**

$$A v = \lambda v$$


* $A$: A square matrix.
* $v$: The eigenvector (direction remains unchanged after transformation).
* $\lambda$: The eigenvalue (scalar scaling factor of the eigenvector).



### Analytical Solution of PCA

Solving the optimization objective function using the Rayleigh Quotient proves that:


$$\Sigma u = \lambda u$$

* The optimal projection vectors $u$ are the **eigenvectors** of the Covariance Matrix $\Sigma$.
* The variance along any eigenvector is equal to its corresponding **eigenvalue** $\lambda$.

---

## 4. Step-by-Step Mathematical Algorithm

Given an input dataset matrix $X$ of shape $(n \times d)$:

### Step 1: Mean Centering

Calculate the mean vector $\mu$ and subtract it from all data points:


$$X_{\text{centered}} = X - \mu$$

### Step 2: Compute Covariance Matrix

Calculate the $d \times d$ covariance matrix $\Sigma$:


$$\Sigma = \frac{1}{n} X_{\text{centered}}^T X_{\text{centered}}$$

### Step 3: Eigen Decomposition

Compute eigenvalues ($\lambda_1, \lambda_2, \dots, \lambda_d$) and eigenvectors ($v_1, v_2, \dots, v_d$) of $\Sigma$:


$$\text{Eigen-Decomposition}(\Sigma) \implies (\lambda_i, v_i)$$

### Step 4: Sort and Select Components

1. Sort eigenvalues in descending order: $\lambda_1 \ge \lambda_2 \ge \dots \ge \lambda_d$.
2. Sort eigenvectors to match the sorted eigenvalues.
3. Select the top $k$ eigenvectors to form a transformation matrix $P$ of shape $(d \times k)$:

$$P = [v_1, v_2, \dots, v_k]$$



### Step 5: Transform the Data Matrix

Project the centered data onto the top $k$ components via matrix multiplication:


$$X_{\text{new}} = X_{\text{centered}} \cdot P$$

* **Dimensions change:** From $(n \times d)$ to $(n \times k)$.

```python
import numpy as np
import pandas as pd

np.random.seed(23) 

mu_vec1 = np.array([0,0,0])
cov_mat1 = np.array([[1,0,0],[0,1,0],[0,0,1]])
class1_sample = np.random.multivariate_normal(mu_vec1, cov_mat1, 20)

df = pd.DataFrame(class1_sample,columns=['feature1','feature2','feature3'])
df['target'] = 1

mu_vec2 = np.array([1,1,1])
cov_mat2 = np.array([[1,0,0],[0,1,0],[0,0,1]])
class2_sample = np.random.multivariate_normal(mu_vec2, cov_mat2, 20)

df1 = pd.DataFrame(class2_sample,columns=['feature1','feature2','feature3'])

df1['target'] = 0

df = df.append(df1,ignore_index=True)

df = df.sample(40)
```

```python
df.head()
```

```python
import plotly.express as px
#y_train_trf = y_train.astype(str)
fig = px.scatter_3d(df, x=df['feature1'], y=df['feature2'], z=df['feature3'],
              color=df['target'].astype('str'))
fig.update_traces(marker=dict(size=12,
                              line=dict(width=2,
                                        color='DarkSlateGrey')),
                  selector=dict(mode='markers'))

fig.show()
     
```

```python
# Step 1 - Apply standard scaling
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

df.iloc[:,0:3] = scaler.fit_transform(df.iloc[:,0:3])
```

```python
# Step 2 - Find Covariance Matrix
covariance_matrix = np.cov([df.iloc[:,0],df.iloc[:,1],df.iloc[:,2]])
print('Covariance Matrix:\n', covariance_matrix)
```

```python
# Step 3 - Finding EV and EVs
eigen_values, eigen_vectors = np.linalg.eig(covariance_matrix)
```

```python
%pylab inline

from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from mpl_toolkits.mplot3d import proj3d
from matplotlib.patches import FancyArrowPatch


class Arrow3D(FancyArrowPatch):
    def __init__(self, xs, ys, zs, *args, **kwargs):
        FancyArrowPatch.__init__(self, (0,0), (0,0), *args, **kwargs)
        self._verts3d = xs, ys, zs

    def draw(self, renderer):
        xs3d, ys3d, zs3d = self._verts3d
        xs, ys, zs = proj3d.proj_transform(xs3d, ys3d, zs3d, renderer.M)
        self.set_positions((xs[0],ys[0]),(xs[1],ys[1]))
        FancyArrowPatch.draw(self, renderer)

fig = plt.figure(figsize=(7,7))
ax = fig.add_subplot(111, projection='3d')

ax.plot(df['feature1'], df['feature2'], df['feature3'], 'o', markersize=8, color='blue', alpha=0.2)
ax.plot([df['feature1'].mean()], [df['feature2'].mean()], [df['feature3'].mean()], 'o', markersize=10, color='red', alpha=0.5)
for v in eigen_vectors.T:
    a = Arrow3D([df['feature1'].mean(), v[0]], [df['feature2'].mean(), v[1]], [df['feature3'].mean(), v[2]], mutation_scale=20, lw=3, arrowstyle="-|>", color="r")
    ax.add_artist(a)
ax.set_xlabel('x_values')
ax.set_ylabel('y_values')
ax.set_zlabel('z_values')

plt.title('Eigenvectors')

plt.show()
```

```python
pc = eigen_vectors[0:2]
pc
```

```python
transformed_df = np.dot(df.iloc[:,0:3],pc.T)
# 40,3 - 3,2
new_df = pd.DataFrame(transformed_df,columns=['PC1','PC2'])
new_df['target'] = df['target'].values
new_df.head()
```
