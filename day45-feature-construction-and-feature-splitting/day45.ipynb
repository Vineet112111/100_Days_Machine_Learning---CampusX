# Feature Construction and Feature Splitting

## Introduction

Feature Engineering is one of the most important steps in Machine Learning.

Two important techniques are:

1. Feature Construction
2. Feature Splitting

These techniques help create better features from existing data, allowing machine learning models to learn more useful patterns.

---

# 1. Feature Construction

## What is Feature Construction?

Feature Construction is the process of creating new features from existing features.

Instead of directly using raw data, we generate more meaningful features that help the model learn better.

---

## Why Do We Need Feature Construction?

Sometimes the original features do not contain enough information.

By combining existing features, we can create more useful information.

Benefits:

- Improves model performance
- Reveals hidden relationships
- Makes patterns easier to learn
- Increases predictive power

---

## Example 1: BMI Calculation

Original Features:

Height = 1.75 m

Weight = 70 kg

Constructed Feature:

BMI = Weight / Height²

BMI = 70 / (1.75)²

BMI = 22.86

Instead of giving Height and Weight separately, BMI provides more meaningful information.

---

## Example 2: Date Feature Construction

Original Feature:

Date = 15-08-2025

Constructed Features:

Day = 15

Month = 8

Year = 2025

Day_of_Week = Friday

Quarter = Q3

Weekend = Yes/No

These new features often improve model performance.

---

## Example 3: House Price Dataset

Original Features:

Length = 20 ft

Width = 30 ft

Constructed Feature:

Area = Length × Width

Area = 600 sq ft

Area is usually more useful than Length and Width separately.

---

## Example 4: Age Calculation

Original Feature:

Date_of_Birth = 01-01-2000

Constructed Feature:

Age = Current Year - Birth Year

Age = 25 years

Age is usually more informative than Date of Birth.

---

# Types of Feature Construction

### Mathematical Construction

Examples:

Area = Length × Width

BMI = Weight / Height²

Speed = Distance / Time

---

### Domain Knowledge Construction

Created using business knowledge.

Examples:

Credit Utilization Ratio

Debt-to-Income Ratio

Customer Lifetime Value

---

### Date-Time Construction

From Date column:

- Day
- Month
- Year
- Weekday
- Weekend
- Quarter

---

### Aggregation-Based Construction

Examples:

Total Purchase Amount

Average Monthly Spending

Total Orders

Maximum Purchase

---

# Advantages of Feature Construction

- Improves model accuracy
- Creates meaningful information
- Captures hidden patterns
- Often more important than choosing algorithms

---

# Disadvantages

- Requires domain knowledge
- Can increase dimensionality
- May create redundant features

---

# 2. Feature Splitting

## What is Feature Splitting?

Feature Splitting means dividing one feature into multiple smaller features.

Sometimes a single column contains multiple pieces of information.

Splitting helps separate those pieces.

---

## Why Use Feature Splitting?

A machine learning model understands structured information better.

Splitting makes hidden information accessible.

---

## Example 1: Full Name

Original Feature:

Name = "Vineet Kumar Yadav"

Split Features:

First_Name = Vineet

Middle_Name = Kumar

Last_Name = Yadav

---

## Example 2: Date

Original Feature:

15-08-2025

Split Features:

Day = 15

Month = 08

Year = 2025

---

## Example 3: Address

Original Feature:

"Lucknow, Uttar Pradesh, India"

Split Features:

City = Lucknow

State = Uttar Pradesh

Country = India

---

## Example 4: Email Address

Original Feature:

vineet@gmail.com

Split Features:

Username = vineet

Domain = gmail.com

Provider = Gmail

---

# Python Example

## Splitting Name

```python
df[['First_Name','Middle_Name','Last_Name']] = df['Name'].str.split(' ', expand=True)
```

---

## Splitting Date

```python
df['Date'] = pd.to_datetime(df['Date'])

df['Day'] = df['Date'].dt.day
df['Month'] = df['Date'].dt.month
df['Year'] = df['Date'].dt.year
```

---

## Splitting Email

```python
df['Username'] = df['Email'].str.split('@').str[0]

df['Domain'] = df['Email'].str.split('@').str[1]
```

---

# Feature Construction vs Feature Splitting

Feature Construction:

Creates new features from existing features.

Example:

Weight + Height → BMI

Length + Width → Area

---

Feature Splitting:

Breaks one feature into multiple features.

Example:

Date → Day + Month + Year

Name → First + Last Name

Address → City + State + Country

---

# Real-World Examples

E-commerce:

Purchase History → Total Spend

---

Banking:

Income & Loan → Debt-to-Income Ratio

---

Healthcare:

Height & Weight → BMI

---

Social Media:

Timestamp → Day, Month, Hour

# Final Revision

Feature Construction:

Existing Features → New Feature

Examples:

Height + Weight → BMI

Length + Width → Area

Date_of_Birth → Age

Purpose:

✓ Add Information
✓ Improve Accuracy

---

Feature Splitting:

One Feature → Multiple Features

Examples:

Date → Day + Month + Year

Name → First + Last Name

Address → City + State + Country

Purpose:

✓ Extract Hidden Information
✓ Improve Learning

Golden Rule:

Feature Construction adds new information.

Feature Splitting extracts information already hidden inside a feature.

````python
import numpy as np
import pandas as pd

from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression

import seaborn as sns
````

````python
df = pd.read_csv('train.csv')[['Age','Pclass','SibSp','Parch','Survived']]
````

````python
df.head()
````

````python
df.dropna(inplace=True)
````

````python
df.head()
````

````python
X = df.iloc[:,0:4]
y = df.iloc[:,-1]
````

````python
X.head()
````

````python
np.mean(cross_val_score(LogisticRegression(),X,y,scoring='accuracy',cv=20))
````

## Applying Feature Construction

````python
X['Family_size'] = X['SibSp'] + X['Parch'] + 1
````

````python
X.head()
````

````python
def myfunc(num):
    if num == 1:
        #alone
        return 0
    elif num >1 and num <=4:
        # small family
        return 1
    else:
        # large family
        return 2
````

````python
myfunc(4)
````

````python
X['Family_type'] = X['Family_size'].apply(myfunc)
````

````python
X.head()
````

````python
X.drop(columns=['SibSp','Parch','Family_size'],inplace=True)
````

````python
X.head()
````

````python
np.mean(cross_val_score(LogisticRegression(),X,y,scoring='accuracy',cv=20))
````
